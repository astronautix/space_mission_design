```python
from datetime import datetime
import msise00

%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
altitidues_km = range(150, 400)
atmos = [ msise00.run(time=datetime(2013, 3, 31, 12), altkm=a, glat=65., glon=-148.) for a in altitidues_km  ]
```

```python
atmos[0]
```

```python
density_O = [atm.O.data[0,0,0,0] for atm in atmos]
density_N2 = [atm.N2.data[0,0,0,0] for atm in atmos]
density_O2 = [atm.O2.data[0,0,0,0] for atm in atmos]
density_Ar = [atm.Ar.data[0,0,0,0] for atm in atmos]
```

```python
plt.figure()
plt.plot(altitidues_km, density_O, label="O")
plt.plot(altitidues_km, density_O2, label="O2")
plt.plot(altitidues_km, density_N2, label="N2")
plt.plot(altitidues_km, density_Ar, label="Ar")

plt.yscale("log")
plt.legend()
```
